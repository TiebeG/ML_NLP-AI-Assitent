```python
import os
import re
import json

INPUT = "../course_materials/discussion_topics.txt"
OUTPUT = "../course_materials/discussion_topics.json"

topics = []

with open(INPUT, "r", encoding="utf-8") as f:
    lines = f.readlines()

pattern = r"(\d+\.\d+)\s+(.*)"

for line in lines:
    line = line.strip()
    if not line:
        continue

    match = re.match(pattern, line)
    if match:
        id_, question = match.groups()
        topics.append({
            "id": id_,
            "question": question.strip()
        })

with open(OUTPUT, "w", encoding="utf-8") as f:
    json.dump(topics, f, indent=4)

print(f"Saved {len(topics)} topics to {OUTPUT}")
```
